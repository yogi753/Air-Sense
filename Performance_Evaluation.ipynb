```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data (replace this with your actual data)
model_names = ["SARIMAX Month Wise", "VAR Month Wise", "PROPHET Month Wise", "LGBM Month Wise", "SARIMAX Week Wise", "VAR week Wise", "PROPHET Week Wise", "LGBM Week Wise"]
rmse_values = [0.453071, 0.628621, 0.6615, 0.1565, 0.209143, 0.619369, 0.6267, 0.0978]
mape_values = [0.383661, 0.553240, 0.4344, 0.1392, 0.153223, 0.527686, 0.4324, 0.0820]
mae_values = [0.409460, 0.517567,  0.5060, 0.1357, 0.180633, 0.500986, 0.5111, 0.0845]
mse_values = [0.205273, 0.395164,  0.437597, 0.024488, 0.043741, 0.383618, 0.392744, 0.009560]
r2_values = [0.37889, 1.654453, 1.43, 0.85, 0.70873, 1.554512, 1.1, 0.94]

data = {
    'Model Names': model_names,
    'RMSE': rmse_values,
    'MAPE': mape_values,
    'MAE': mae_values,
    'MSE': mse_values,
    'R2': r2_values
}

df = pd.DataFrame(data)

# Set the figure size and create a bar chart
plt.figure(figsize=(20, 8))

# Define the number of bars and their positions
num_bars = len(df)
bar_width = 0.14
index = np.arange(num_bars)

# Define colors for each metric
colors = {'RMSE': 'lightgreen', 'MAPE': 'orange', 'MAE': 'pink', 'MSE': 'lightblue','R2':'grey',}

# Loop through each metric and plot it
for i, (metric, color) in enumerate(colors.items()):
    plt.bar(index + i * bar_width, df[metric], width=bar_width, label=metric, color=color)

# Customize the plot
plt.xlabel('Algorithms')
plt.ylabel('Value')
plt.title('Performance of Forecasting Models in Month and Week wise')
plt.xticks(index + bar_width * 1.5, df['Model Names'])
plt.legend(loc='best')

# Add data labels above each bar
for i, metric in enumerate(colors.keys()):
    for j, value in enumerate(df[metric]):
        plt.text(index[j] + i * bar_width, value + 0.01 , f'{value:.2f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

**R Square**

```python
import matplotlib.pyplot as plt

# Sample data (replace this with your actual data)
model_names = ["SARIMAX Month", "VAR Month", "PROPHET Month", "LGBM Month", "SARIMAX Week", "VAR Week", "PROPHET Week", "LGBM Week"]
model_types = ["Month Wise", "Month Wise", "Month Wise", "Month Wise", "Week Wise", "Week Wise", "Week Wise", "Week Wise"]
r2_values = [0.37889, 1.654453, 1.43, 0.85, 0.70873, 1.554512, 1.1, 0.94]

# Define colors for "Month Wise" and "Week Wise"
colors = ['skyblue' if model_type == "Month Wise" else 'palegreen' for model_type in model_types]

# Define a unique color for "LGBM Month" and "LGBM Week" bars
lgbm_color = 'pink'  # You can change this to your desired color

# Create a bar chart for R-squared values with different colors
plt.figure(figsize=(8, 6))
bars = plt.bar(model_names, r2_values, width=0.3, color=colors)

# Set a different color for "LGBM Month" and "LGBM Week" bars
for i, model_name in enumerate(model_names):
    if "LGBM" in model_name:
        bars[i].set_color(lgbm_color)

plt.xlabel('Algorithms')
plt.ylabel('R-Square (R2)')

# Set the title with a thicker border (outline)
plt.title('Comparison chart for R-Square (R2) Metric', pad=20)

# Rotate x-axis labels for better alignment
plt.xticks(rotation=45, ha="right")

# Adjust spacing to prevent labels from being cut off
plt.tight_layout()

# Add R2 values above each bar
for bar, r2 in zip(bars, r2_values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.01, f'{r2:.2f}', ha='center', va='bottom')

# Add labels for "Month Wise" and "Week Wise"
month_wise_patch = plt.Rectangle((0, 0), 1, 1, fc="skyblue", edgecolor="none")
week_wise_patch = plt.Rectangle((0, 0), 1, 1, fc="palegreen", edgecolor="none")
lgbm_patch = plt.Rectangle((0, 0), 1, 1, fc=lgbm_color, edgecolor="none")

plt.legend([month_wise_patch, week_wise_patch, lgbm_patch], ['Month Wise', 'Week Wise', 'LGBM'])

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data (replace this with your actual data)
model_names = ["SARIMAX Month", "VAR Month", "PROPHET Month", "LGBM Month", "SARIMAX Week", "VAR Week", "PROPHET Week", "LGBM Week"]
model_types = ["Month Wise", "Month Wise", "Month Wise", "Month Wise", "Week Wise", "Week Wise", "Week Wise", "Week Wise"]
mse_values = [0.205273, 0.395164,  0.437597, 0.024488, 0.043741, 0.383618, 0.392744, 0.009560]

# Define colors for "Month Wise" and "Week Wise"
colors = ['skyblue' if model_type == "Month Wise" else 'palegreen' for model_type in model_types]

# Define a unique color for "LGBM Month" and "LGBM Week" bars
lgbm_color = 'pink'  # You can change this to your desired color

# Create a bar chart for MSE values with different colors
plt.figure(figsize=(8, 6))
bars = plt.bar(model_names, mse_values, width=0.3, color=colors)

# Set a different color for "LGBM Month" and "LGBM Week" bars
for i, model_name in enumerate(model_names):
    if "LGBM" in model_name:
        bars[i].set_color(lgbm_color)

plt.xlabel('Algorithms')
plt.ylabel('MSE (Mean Squared Error)')

# Set the title with a thicker border (outline)
plt.title('Comparison chart for MSE (Mean Squared Error) Metric', pad=20)

# Rotate x-axis labels for better alignment
plt.xticks(rotation=45, ha="right")

# Set the y-axis limits to cover the range [0, 1.6]
plt.ylim(0, 1.6)

# Adjust spacing to prevent labels from being cut off
plt.tight_layout()

# Add MSE values above each bar
for bar, mse in zip(bars, mse_values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.01, f'{mse:.3f}', ha='center', va='bottom')

# Add labels for "Month Wise" and "Week Wise"
month_wise_patch = plt.Rectangle((0, 0), 1, 1, fc="skyblue", edgecolor="none")
week_wise_patch = plt.Rectangle((0, 0), 1, 1, fc="palegreen", edgecolor="none")
lgbm_patch = plt.Rectangle((0, 0), 1, 1, fc=lgbm_color, edgecolor="none")

plt.legend([month_wise_patch, week_wise_patch, lgbm_patch], ['Month Wise', 'Week Wise', 'LGBM'])

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the years and corresponding R-squared values for each model
years = list(range(2015, 2021))
model1_r2 = [0.85, 0.88, 0.82, 0.90, 0.87, 0.89, 0.86]  # Example R-squared values for Model 1
model2_r2 = [0.87, 0.89, 0.86, 0.88, 0.85, 0.87, 0.89]  # Example R-squared values for Model 2
model3_r2 = [0.94, 0.93, 0.95, 0.92, 0.96, 0.94, 0.91]  # Example R-squared values for Model 3

# Calculate average R-squared values for each year
avg_r2 = [np.mean([model1_r2[i], model2_r2[i], model3_r2[i]]) for i in range(len(years))]

# Calculate deviations from the average for each model
model1_deviations = [r2 - avg for r2, avg in zip(model1_r2, avg_r2)]
model2_deviations = [r2 - avg for r2, avg in zip(model2_r2, avg_r2)]
model3_deviations = [r2 - avg for r2, avg in zip(model3_r2, avg_r2)]

# Create the line plots for deviations of each model
plt.plot(years, model1_deviations, marker='o', linestyle='-', color='red', label='Model 1 Deviations')
plt.plot(years, model2_deviations, marker='o', linestyle='-', color='blue', label='Model 2 Deviations')
plt.plot(years, model3_deviations, marker='o', linestyle='-', color='green', label='Model 3 Deviations')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Deviations from Average R-squared')
plt.title('Deviations in R-squared Values Over the Years')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
